```python
import pandas as pd
```

```python
df = pd.read_csv('global_freelancers_raw.csv')
df.columns = df.columns.str.capitalize()

# Standardizing 'Gender' Column.
df["Gender"] = df["Gender"].replace(['f','F','female','FEMALE'],["Female","Female","Female","Female"])
df["Gender"] = df["Gender"].replace(['m','M','male','MALE'],["Male","Male","Male","Male"])

# Dropping rows with missing values in specific columns.
df = df.dropna(subset=['Years_of_experience', 'Hourly_rate (usd)'], how='all')
df = df.dropna(subset=['Rating', 'Hourly_rate (usd)'], how='all')
df = df.dropna(subset=['Client_satisfaction', 'Hourly_rate (usd)'], how='all')
df = df.dropna(subset=['Rating', 'Is_active'], how='all')
df = df.dropna(subset=['Rating', 'Client_satisfaction'], how='all')

# Standardizing 'Hourly_rate (usd)' Column.
df["Hourly_rate (usd)"] = df["Hourly_rate (usd)"].str.replace('USD', '').str.strip()
df["Hourly_rate (usd)"] = df["Hourly_rate (usd)"].str.replace('$', '').str.strip()

# Cleaning 'Is_active' Column.
df["Is_active"] = df["Is_active"].replace(['yes', 'Yes', 'YES','1',1,'y','Y','true','True'], ['Yes','Yes','Yes','Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes'])
df["Is_active"] = df["Is_active"].replace(['no', 'No', 'NO', '0', 0, 'n', 'N','false','False'], ['No','No','No', 'No', 'No', 'No', 'No', 'No', 'No'])

df['Client_satisfaction'] = df['Client_satisfaction'].apply(lambda x :x if pd.isna(x) or x.endswith('%') else x + '%')

# Saving the cleaned DataFrame to a new CSV file.
df.to_csv('global_freelancers_cleaned.csv', index=False)
```
